```python
import pandas as pd
df = pd.read_csv('datasets/heart-disease.csv')

df.head()
```

```python
x=df.drop(columns=['target'])
y=df['target']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
lr=LogisticRegression()
cls=DecisionTreeClassifier()
lr.fit(x_train,y_train)
cls.fit(x_train,y_train)
lr_pred=lr.predict(x_test)
cls_pred=cls.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix,recall_score,precision_score,f1_score
print('logistic regression: ', accuracy_score(y_test,lr_pred))
print('Decision tree classifieer: ', accuracy_score(y_test,cls_pred))
```

```python

print("logistic Regresion:   ")
print("accuracy_score:     ",accuracy_score(y_test,lr_pred))
print("recall_score:       ",recall_score(y_test,lr_pred))
print("precision_score:    ",precision_score(y_test,lr_pred))
print("f1_score:           ",f1_score(y_test,lr_pred))
print("confusion_matrix:   \n",confusion_matrix(y_test,lr_pred))
print("--------------------------------------------------------")
print()
print("Decision tree classifier:   ")
print("accuracy_score:   ",accuracy_score(y_test,cls_pred))
print("recall_score:   ",recall_score(y_test,cls_pred))
print("precision_score:   ",precision_score(y_test,cls_pred))
print("f1_score:   ",f1_score(y_test,cls_pred))
print("confusion_matrix:  \n",confusion_matrix(y_test,cls_pred))
```
