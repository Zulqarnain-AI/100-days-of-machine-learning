```python
import pandas as pd 
import seaborn as sns

```

```python
df=sns.load_dataset('titanic')
```

```python
df.head()
```

```python
x=df['class']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
x=pd.DataFrame(x)
```

```python
oe=OrdinalEncoder(categories=[['First','Second','Third']])
```

```python
oe.fit(x)
```

```python
df['class']=oe.transform(x)
```

```python
df.head()
```
