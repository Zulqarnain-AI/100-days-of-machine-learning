```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('datasets/student_clustering.csv')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(df['cgpa'],df['iq'])
```

```python
from sklearn.cluster import KMeans
```

```python
wcss = []
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(df)
    wcss.append(km.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
x = df.iloc[:,:].values
```

```python
km=KMeans(n_clusters=4)
y_mean= km.fit_predict(x)
```

```python
x[y_mean==0]
```

```python
plt.scatter(x[y_mean==0,0],x[y_mean==0,1],color='red')
plt.scatter(x[y_mean==1,0],x[y_mean==1,1],color='green')
plt.scatter(x[y_mean==2,0],x[y_mean==2,1],color='blue')
plt.scatter(x[y_mean==3,0],x[y_mean==3,1],color='black')
```
