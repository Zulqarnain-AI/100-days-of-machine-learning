```python
import pickle
import numpy as np
import pandas as pd
```

```python
oheSex=pickle.load(open('trained_models/oheSex.pkl','rb'))
oheEmbarked=pickle.load(open('trained_models/oheEmbarked.pkl','rb'))
clf=pickle.load(open('trained_models/clf.pkl','rb'))
```

```python
# pclass,sex,age,sibsp,fare,embarked 
input=np.array([3,'female',31,0,1,20,'S'],dtype=object).reshape(1,7)
```

```python
input
```

```python
input_sex=oheSex.transform(input[:,1].reshape(1,1))
```

```python
input_sex
```

```python
input_embarked=oheEmbarked.transform(input[:,-1].reshape(1,1))
```

```python
input_embarked
```

```python
input_age=input[:,2].reshape(1,1)
```

```python
input
```

```python
remaining_input=input[:,[0,3,4,5]]
```

```python
remaining_input
```

```python
input_transformed=np.concatenate((remaining_input,input_age,input_sex,input_embarked),axis=1)
```

```python
input_transformed
```

```python
clf.predict(input_transformed)
```
