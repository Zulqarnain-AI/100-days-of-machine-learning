```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
```

```python
df=sns.load_dataset('titanic')
```

```python
df=df.drop(columns=['pclass','sex','parch','embarked','class','who','adult_male','deck','embark_town','alive','alone'])
```

```python
df.isnull().mean()
```

```python
x=df.drop(columns=['survived'])
y=df['survived']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=2)
```

```python
x_train.shape,x_test.shape
```

```python
x_train.isnull().mean()
```

```python
mean_age=x['age'].mean()
median_age=x['age'].median()

mean_fare=x['age'].mean()
median_fare=x['age'].median()
```

```python
x_train['age_mean']=x_train['age'].fillna(mean_age)
x_train['age_median']=x_train['age'].fillna(median_age)

x_train['fare_median']=x_train['fare'].fillna(median_fare)
x_train['fare_mean']=x_train['fare'].fillna(mean_fare)

x_test['age_mean']=x_test['age'].fillna(mean_age)
x_test['age_median']=x_test['age'].fillna(median_age)

x_test['fare_median']=x_test['fare'].fillna(median_fare)
x_test['fare_mean']=x_test['fare'].fillna(mean_fare)
```

```python
x_train.sample(5)
```

```python
print('age before impute', x_train['age'].var())
print('age after mean imputation', x_train['age_mean'].var())
print('age after median imputation', x_train['age_median'].var())

print()
print('fare before impute', x_train['fare'].var())
print('fare after mean imputation', x_train['fare_mean'].var())
print('fare after median imputation', x_train['fare_median'].var())
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

# origina, distribution
x_train['age'].plot(kind='kde',ax=ax)

# after imputation
x_train['age_mean'].plot(kind='kde',ax=ax,color='red')
x_train['age_median'].plot(kind='kde',ax=ax,color='green')

# add legends
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

```python
fig=plt.figure()
ax=fig.add_subplot(111)

# origina, distribution
x_train['fare'].plot(kind='kde',ax=ax)

# after imputation
x_train['fare_mean'].plot(kind='kde',ax=ax,color='red')
x_train['fare_median'].plot(kind='kde',ax=ax,color='green')

# add legends
lines,labels=ax.get_legend_handles_labels()
ax.legend(lines,labels,loc='best')
```

```python
imputer1=SimpleImputer(strategy='median')
imputer2=SimpleImputer(strategy='mean')
```

```python
trf=ColumnTransformer([
    ('imputer1',imputer1,['age']),
    ('imputer2',imputer2,['fare'])],remainder='passthrough')
```

```python
x_train
```

```python
trf.fit(x_train)
```

```python
x_train=trf.transform(x_train)
x_test=trf.transform(x_test)
```
