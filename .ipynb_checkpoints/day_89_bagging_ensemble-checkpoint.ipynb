```python
import pandas as pd
import numpy as np
import sklearn
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder,OrdinalEncoder
from sklearn.pipeline import Pipeline
from sklearn.metrics import accuracy_score,confusion_matrix,precision_score,recall_score,f1_score
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
```

```python
df=pd.read_csv('datasets/fyp_data.csv')
```

```python
df=df[['Symptoms','Age','Sex','Nature','Disease']]

df['Symptoms'] = df['Symptoms'].fillna('Unknown')
df['Age'] = df['Age'].fillna(df['Age'].median())
df['Sex'] = df['Sex'].fillna('unknown')
df['Nature'] = df['Nature'].fillna('Unknown')

x_train,x_test,y_train,y_test=train_test_split(df.drop(columns=['Disease']),df['Disease'],test_size=0.2,random_state=42)

cate=['Symptoms','Sex']
order=['Nature']

processing=ColumnTransformer(
    transformers=[
        ('categorical_encoding',OneHotEncoder(sparse_output=False),cate),
        ('ordinal_encoding',OrdinalEncoder(categories=[['Unknown','medium','high']]),order)
    ],
    remainder='passthrough'
)

pip=Pipeline(
    steps=[
        ('preprocess',processing)
    ]
)

x_train_trf=pip.fit_transform(x_train)
x_test_trf=pip.fit_transform(x_test)

from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
y_train_trf=le.fit_transform(y_train)
y_test_trf=le.fit_transform(y_test)
```

```python
dtc=DecisionTreeClassifier(max_depth=10)
dtc.fit(x_train_trf,y_train)
y_pred=dtc.predict(x_test_trf)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.ensemble import BaggingClassifier
dtb=BaggingClassifier(
    estimator=DecisionTreeClassifier(max_depth=10),
    n_estimators=50,
    max_samples=0.2,
    bootstrap=True,
    random_state=42
)
dtb.fit(x_train_trf,y_train_trf)
yg_pred=dtb.predict(x_test_trf)
accuracy_score(y_test_trf,yg_pred)
```

```python
dtk=BaggingClassifier(
    estimator=KNeighborsClassifier(),
    n_estimators=50,
    max_samples=0.2,
    bootstrap=True,
    random_state=42
)
dtk.fit(x_train_trf,y_train_trf)
yk_pred=dtk.predict(x_test_trf)
accuracy_score(y_test_trf,yk_pred)
```

```python
dtr=BaggingClassifier(
    estimator=RandomForestClassifier(),
    n_estimators=50,
    max_samples=0.2,
    bootstrap=True,
    random_state=42
)
dtr.fit(x_train_trf,y_train_trf)
yr_pred=dtr.predict(x_test_trf)
accuracy_score(y_test_trf,yr_pred)
```
