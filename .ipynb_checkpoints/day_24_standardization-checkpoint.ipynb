```python
x=df.drop(['survived','sex','embarked','class','who','adult_male','deck','embark_town','alive','alone'],axis=1)
y=df['survived']
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
```

```python
ss.fit(x)
```

```python
scaled_x=ss.transform(x)
```

```python
sxd=pd.DataFrame(scaled_x,columns=x.columns)
```

```python
sxd.head()
```

```python
x.head(2)
```

```python
import matplotlib.pyplot as olt
import seaborn as sns
```

```python
fig,(f1,f2)=olt.subplots(ncols=2,figsize=(15,10))

f1.scatter(x['age'],x['fare'])
f1.set_title("before scaling")

f2.scatter(sxd['age'],sxd['fare'],color='red')
f2.set_title("after scaling")

olt.show()
```

```python
fig,(f1,f2)=olt.subplots(ncols=2,figsize=(10,5))

f1.set_title("before ")
sns.kdeplot(x['age'],ax=f1)
sns.kdeplot(x['fare'],ax=f1)

f2.set_title("after ")
sns.kdeplot(sxd['age'],ax=f2)
sns.kdeplot(sxd['fare'],ax=f2)

olt.show()
```
